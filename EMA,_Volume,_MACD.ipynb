```python
reli = pd.read_csv('RELIANCE.NS.csv')
reli = reli.set_index(pd.DatetimeIndex(reli['Date']))
```

#Analyze RELIANCE Stock


```python
reli.head()
```

```python
print(reli)
```

```python
reli['PriceDiff'] = reli['Close'].shift(-1) - reli['Close']
reli['DailyReturn'] =reli['PriceDiff'] / reli['Close']
print(reli)
```

#MOVING AVERAGES: 5, 13, 26

```python
#daily returns
reli['Direction'] = [1 if reli['DailyReturn'].loc[ei] > 0 else 0 for ei in reli.index]
```

```python
graph= {
        'x': reli.Date,
        'open': reli.Open,
        'close': reli.Close,
        'low': reli.Low,
        'high': reli.High,
        'type' :'candlestick'              
}
```

```python
reli['MA5'] = reli['Close'].rolling(window=3, min_periods=1).mean()
reli['MA13'] = reli['Close'].rolling(window=13, min_periods=1).mean()
reli['MA26'] = reli['Close'].rolling(window=26, min_periods=1).mean()

avg1= {
    'x': reli.Date,
    'y': reli.MA5,
    'type':'scatter',
    'mode':'lines',
    'line': {
        'width':1,
        'color': 'black'
    },
    'name':'Moving Average of 5 periods'

}

avg2= {
    'x': reli.Date,
    'y': reli.MA13,
    'type':'scatter',
    'mode':'lines',
    'line': {
        'width':1,
        'color': 'purple'
    },
    'name':'Moving Average of 13 periods'

}

avg3= {
    'x': reli.Date,
    'y': reli.MA26,
    'type':'scatter',
    'mode':'lines',
    'line': {
        'width':1,
        'color': 'orange'
    },
    'name':'Moving Average of 26 periods'

}



data=[graph, avg1, avg2, avg3]

layout= go.Layout({
    'title':{
        'text': 'RELIANCE NS (2011-2021)',
        'font': {
            'size':25
            
        }
    }
}
)
```

```python
fig= go.Figure(data=data, layout=layout)
config = dict({'scrollZoom': True})
fig.show(config=config)
```

# VOLUME


```python
fig = make_subplots(specs=[[{"secondary_y": True}]])


fig.add_trace(go.Candlestick(x=reli['Date'],
                open=reli['Open'], high=reli['High'],
                low=reli['Low'], close=reli['Close']),
               secondary_y=True)

# include a go.Bar trace for volumes
fig.add_trace(go.Bar(x=reli['Date'], y=reli['Volume']),
               secondary_y=False)

fig.layout.yaxis2.showgrid=False
config = dict({'scrollZoom': True})
fig.show(config=config)
```

#MACD SIGNAL

```python
reli['MA12'] = reli['Close'].rolling(window=12, min_periods=1).mean()
reli['MA26'] = reli['Close'].rolling(window=26, min_periods=1).mean()
reli['macd'] = reli['MA12'] - reli['MA26']
reli['exp3'] = reli['macd'].rolling(window=9, min_periods=1).mean()
```

```python
MACD1= {
    'x': reli.Date,
    'y': reli.macd,
    'type':'scatter',
    'mode':'lines',
    'line': {
        'width':1,
        'color': 'green'
    },
    'name':'MACD line'

}
```

```python
signal= {
    'x': reli.Date,
    'y': reli.exp3,
    'type':'scatter',
    'mode':'lines',
    'line': {
        'width':1,
        'color': 'red'
    },
    'name':'Signal Line'

}
```

```python
data=[MACD1,signal]

layout= go.Layout({
    'title':{
        'text': 'RELIANCE NS (2011-2021) MACD & Signal',
        'font': {
            'size':25
            
        }
    }
}
)
```

```python
fig= go.Figure(data=data, layout=layout)
fig.update_layout(xaxis_rangeslider_visible=True)
config = dict({'scrollZoom': True})
fig.show(config=config)
```
